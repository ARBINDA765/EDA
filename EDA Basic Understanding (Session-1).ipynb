## Univariate,Bivariate and MultiVariate Analysis

```python
#Importing the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading the data 
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

### Univariate Analysis

```python
#Creating the dataframe where species'=='setosa'
df_setosa=df.loc[df['species']=='setosa']
```

```python
#Creating the dataframe where species'=='virginica'
df_virginica=df.loc[df['species']=='virginica']
```

```python
#Creating the dataframe where species'=='versicolor'
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'*',color='blue',linestyle='dashed',linewidth=3, markersize=12)
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'*',color='green',linestyle='dashed',linewidth=3, markersize=12)
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'*',color='red',linestyle='dashed',linewidth=3, markersize=12)
plt.xlabel('SePal length')
plt.show()
```

### Bivariate Analysis

```python
sns.FacetGrid(df,hue="species",size=7,height=9).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```
